```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# updated dataframe - from April 2018 to June 2019

mask_new = (df2_all_full['Date'] >= '2020-11-01') & (df2_all_full['Date'] <= '2021-11-30')

df2_all_updated = df2_all_full.loc[mask_new]
```

```python
# Share of categories

sns.set(style="whitegrid")

colors = ['b', 'g']
sns.set_palette(colors)

plt.figure(figsize=(10,8))
total = len(df2_all_updated)
ax = sns.countplot(x="category", data=df2_all_updated)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
plt.xlabel('Category', size = 12)
plt.ylabel('Number of tweets', size = 12)
plt.ylim(0,7500)
plt.title('Dataset 1')
plt.show()
```

```python
# share of sentiment

colors = ['b', 'g', 'r']
sns.set_palette(colors)

plt.figure(figsize=(10,8))
total = len(df2_all_updated)
ax = sns.countplot(x="sentiment", data=df2_all_updated)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
plt.xlabel('Sentiment', size = 12)
plt.ylabel('Number of tweets', size = 12)
plt.ylim(0,4000)
plt.title('Dataset 1')
plt.show()
```

```python
# distribution of sentiment by category

# count plot on two categorical variable
fig, ax = plt.subplots(figsize=(12, 6))
ax1 = sns.countplot(x ='sentiment', hue = "category", data = df2_all_updated, ax=ax)
plt.xticks(rotation=0)
plt.ylabel('Count')
plt.xlabel('Sentiment')

for container in ax1.containers:
    ax1.bar_label(container)

    
plt.show()
```

```python
# add a month year column to the monthly dataframe

df2_all_monthly['month_year'] = df2_all_monthly['Date'].dt.to_period('M')
```

```python
# create separate dataframes from with and without games
df2_all_monthly_only_games = df2_all_monthly.loc[[2,6,9,12,13,14], :]      
df2_all_monthly_no_games = df2_all_monthly.loc[[3,4,5,7,8,10,11], :]
```

```python
# calculating totals of each column in the dataframes 
df2_all_monthly_only_games.loc["Matches"] = df2_all_monthly_only_games.sum().astype(np.int64)
df2_all_monthly_no_games.loc["No Matches"] = df2_all_monthly_no_games.sum().astype(np.int64)
```

```python
# create dataframes with only totals of both dataframes
totals_df2 = pd.concat([df2_all_monthly_only_games[-1:-2:-1],df2_all_monthly_no_games[-1:-2:-1]])

# calculating percentages of each sentiment
totals_df2['Negative-Political'] = (totals_df2['Negative Political'] /(totals_df2['Negative Political']+totals_df2['Negative Professional']+
                                                               totals_df2['Neutral Political']+totals_df2['Neutral Professional']+
                                                             totals_df2['Positive Political']+totals_df2['Positive Professional']))*100

totals_df2['Negative-Professional'] = (totals_df2['Negative Professional'] /(totals_df2['Negative Political']+totals_df2['Negative Professional']+
                                                               totals_df2['Neutral Political']+totals_df2['Neutral Professional']+
                                                             totals_df2['Positive Political']+totals_df2['Positive Professional']))*100

totals_df2['Neutral-Political'] = (totals_df2['Neutral Political'] /(totals_df2['Negative Political']+totals_df2['Negative Professional']+
                                                               totals_df2['Neutral Political']+totals_df2['Neutral Professional']+
                                                             totals_df2['Positive Political']+totals_df2['Positive Professional']))*100

totals_df2['Neutral-Professional'] = (totals_df2['Neutral Professional'] /(totals_df2['Negative Political']+totals_df2['Negative Professional']+
                                                               totals_df2['Neutral Political']+totals_df2['Neutral Professional']+
                                                             totals_df2['Positive Political']+totals_df2['Positive Professional']))*100

totals_df2['Positive-Political'] = (totals_df2['Positive Political'] /(totals_df2['Negative Political']+totals_df2['Negative Professional']+
                                                               totals_df2['Neutral Political']+totals_df2['Neutral Professional']+
                                                             totals_df2['Positive Political']+totals_df2['Positive Professional']))*100

totals_df2['Positive-Professional'] = (totals_df2['Positive Professional'] /(totals_df2['Negative Political']+totals_df2['Negative Professional']+
                                                               totals_df2['Neutral Political']+totals_df2['Neutral Professional']+
                                                             totals_df2['Positive Political']+totals_df2['Positive Professional']))*100

```

```python
totals_df2.reset_index(inplace=True)
```

```python
# sentiment share in months and no months - total 

fig, ax = plt.subplots(figsize=(12, 6))

ax1 = totals_df2.plot.bar(x='index',
                                y=['Neutral-Professional','Positive-Professional',"Negative-Professional",
                                    "Neutral-Political", 'Positive-Political',"Negative-Political"],
                                      color=["purple", 'green', 'blue', 'red', 'brown', 'orange'],        
                                      stacked=True, ax = ax)         
                                  

plt.xlabel('All months - aggregated')
plt.ylabel('%')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.23, 1.02 ))
plt.show()
```
